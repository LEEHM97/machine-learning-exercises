## 스케일링

```python
# multiple_model 만 범주형 데이터

categorical_df = df[['multiple_model']]
numerical_df = df.drop(columns=['multiple_model'])
numerical_df
```

```python
# 수치형 데이터 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scalied_df = numerical_df.copy()

scalied_df = scaler.fit_transform(scalied_df)
numerical_df = pd.DataFrame(scalied_df, columns=numerical_df.columns)
numerical_df.head()
```

```python
df = numerical_df.join(categorical_df)
df
```
