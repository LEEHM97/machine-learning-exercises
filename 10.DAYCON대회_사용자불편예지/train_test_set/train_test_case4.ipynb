# 스케일링

## train data

범주형 - 수치형 데이터 나누기

```python
# multiple_model 만 범주형 데이터

train_categorical = train[['multiple_model']]
train_numerical = train.drop(columns=['multiple_model'])

test_categorical = test[['multiple_model']]
test_numerical = test.drop(columns=['multiple_model'])
```

```python
# 수치형 데이터 스케일링
# train이기 때문에 fit-transform
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()   
scaler.fit(train_numerical)

scalied_train = train_numerical.copy()
scalied_test = train_numerical.copy()
```

```python
scalied_train = scaler.transform(train_numerical)
scalied_test = scaler.transform(test_numerical)
```

```python
train_numerical = pd.DataFrame(scalied_train, columns=train_numerical.columns)
test_numerical = pd.DataFrame(scalied_test, columns=test_numerical.columns)

display(train_numerical.head())
display(test_numerical.head())
```

```python
train = train_numerical.join(train_categorical)
test = test_numerical.join(test_categorical)
```

```python
train.to_csv('../training set/train_case4.csv', index=False)
test.to_csv('../training set/test_case4.csv', index=False)
```
